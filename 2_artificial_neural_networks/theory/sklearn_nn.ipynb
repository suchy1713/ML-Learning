```
import numpy as np
import matplotlib.pyplot as plt

#generate random data
N = 500
D = 2 # dimensionality of input
M = 10 # hidden layer size
K = 3 # number of classes

X1 = np.random.randn(N, 2) + np.array([0, -2])
X2 = np.random.randn(N, 2) + np.array([2, 2])
X3 = np.random.randn(N, 2) + np.array([-2, 2])
X = np.vstack([X1, X2, X3])

Y = np.array([0]*N + [1]*N + [2]*N)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=100, alpha=0.5)
plt.show()

N = len(Y)
T = np.zeros((N, K))
for i in range(N):
    T[i, Y[i]] = 1
```

```
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

X, T = shuffle(X, T)
X_train, X_test, Y_train, Y_test = train_test_split(X, T, test_size=0.2)

brain = MLPClassifier(hidden_layer_sizes=(10,),
                      learning_rate_init=0.05,
                      max_iter=1000)

brain.fit(X_train, Y_train)

print('Train accuracy: ', brain.score(X_train, Y_train))
print('Test accuracy: ', brain.score(X_test, Y_test))
```
