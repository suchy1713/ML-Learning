```
import numpy as np
import matplotlib.pyplot as plt

#generate random data
n = 500
D = 2 # dimensionality of input
K = 3 # number of classes

X1 = np.random.randn(n, 2) + np.array([0, -2])
X2 = np.random.randn(n, 2) + np.array([2, 2])
X3 = np.random.randn(n, 2) + np.array([-2, 2])
X = np.vstack([X1, X2, X3])

Y = np.array([0]*n + [1]*n + [2]*n)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=100, alpha=0.5)
plt.show()

N = len(Y)
Y_ind = np.zeros((N, K))
for i in range(N):
    Y_ind[i, Y[i]] = 1
```

```
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

from keras.models import Sequential, Model
from keras.layers import Dense, Activation, Input

def one_way(K, D, X, Y_ind):
    brain = Sequential()

    brain.add(Dense(units=10, input_dim=D))
    brain.add(Activation('relu'))
    brain.add(Dense(units=K))
    brain.add(Activation('softmax'))

    brain.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )

    output = brain.fit(X, Y_ind, validation_split=0.33, epochs=50)

    plt.plot(output.history['loss'], label='Train loss')
    plt.plot(output.history['val_loss'], label='Validation loss')
    plt.legend()
    plt.show()

    plt.plot(output.history['acc'], label='Train accuracy')
    plt.plot(output.history['val_acc'], label='Validation accuracy')
    plt.legend()
    plt.show()


def another_way(K, D, X, Y_ind):
    i = Input(shape=(D,))
    x = Dense(10, activation='relu')(i)
    x = Dense(K, activation='softmax')(x)
    brain = Model(inputs=i, outputs=x)

    brain.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )

    output = brain.fit(X, Y_ind, validation_split=0.33, epochs=50)

    plt.plot(output.history['loss'], label='Train loss')
    plt.plot(output.history['val_loss'], label='Validation loss')
    plt.legend()
    plt.show()

    plt.plot(output.history['acc'], label='Train accuracy')
    plt.plot(output.history['val_acc'], label='Validation accuracy')
    plt.legend()
    plt.show()

#one_way(K, D, X, Y_ind)
another_way(K, D, X, Y_ind)
```
