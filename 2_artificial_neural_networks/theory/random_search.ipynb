```
from google.colab import drive 
drive.mount('/content/gdrive')
import pandas as pd
pd.set_option('display.max_columns', 500)
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('gdrive/My Drive/csvs/Iris.csv')

scaler = StandardScaler()

df['SepalLengthCm'] = scaler.fit_transform(df[['SepalLengthCm']].values.astype(float))
df['SepalWidthCm'] = scaler.fit_transform(df[['SepalWidthCm']].values.astype(float))
df['PetalLengthCm'] = scaler.fit_transform(df[['PetalLengthCm']].values.astype(float))
df['PetalWidthCm'] = scaler.fit_transform(df[['PetalWidthCm']].values.astype(float))

df = pd.get_dummies(df, columns=['Species'])

print(df.head())
```

```
from sklearn.utils import shuffle

X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
Y = df[['Species_Iris-setosa', 'Species_Iris-versicolor', 'Species_Iris-virginica']].values
X, Y = shuffle(X, Y)

from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
import numpy as np

log_lr = -4
log_r = -2
log_mu = -2
n_hidden = 1
size_hidden = 6

score = 0
f_lr, f_r, f_nh, f_m, f_mu = 0, 0, 0, 0, 0

#Random Search
for i in range(50):
  brain = MLPClassifier(
      hidden_layer_sizes=(tuple([size_hidden]*n_hidden)),
      learning_rate_init=10**log_lr,
      momentum=1-10**log_mu,
      alpha=10**log_r,
      max_iter=100000
  )

  print(i, '/50')
  new_score = np.mean(cross_val_score(brain, X, Y, cv=3))
  if new_score > score:
    score = new_score
    f_lr = log_lr
    f_mu = log_mu
    f_r = log_r
    f_nh = n_hidden
    f_m = size_hidden

  log_lr = f_lr + np.random.randint(-1, 2)
  log_r = f_r + np.random.randint(-1, 2)
  log_mu = f_mu + np.random.randint(-1, 2)
  n_hidden = max(1, f_nh + np.random.randint(-1, 2))
  size_hidden = max(2, f_m + np.random.randint(-3, 6))

print('Learning rate: ', 10**f_lr)
print('Momentum: ', 1-10**f_mu)
print('Regularization: ', 10**f_r)
print('Hidden layers: ', tuple([f_m]*f_nh))
print('Score: ', score)
```
