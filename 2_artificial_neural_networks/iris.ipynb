```
from google.colab import drive 
drive.mount('/content/gdrive')
import pandas as pd
pd.set_option('display.max_columns', 500)
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('gdrive/My Drive/csvs/Iris.csv')

scaler = StandardScaler()

df['SepalLengthCm'] = scaler.fit_transform(df[['SepalLengthCm']].values.astype(float))
df['SepalWidthCm'] = scaler.fit_transform(df[['SepalWidthCm']].values.astype(float))
df['PetalLengthCm'] = scaler.fit_transform(df[['PetalLengthCm']].values.astype(float))
df['PetalWidthCm'] = scaler.fit_transform(df[['PetalWidthCm']].values.astype(float))

df = pd.get_dummies(df, columns=['Species'])

print(df.head())
```

```
from sklearn.utils import shuffle

X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
Y = df[['Species_Iris-setosa', 'Species_Iris-versicolor', 'Species_Iris-virginica']].values
X, Y = shuffle(X, Y)

from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
import numpy as np

learning_rate = [10e-1, 10e-2, 10e-3, 10e-4, 10e-5, 10e-6, 10e-7]
momentum = [1-10e-1, 1-10e-2, 1-10e-3, 1-10e-4]
regularization = [10e-1, 10e-2, 10e-3, 10e-4, 10e-5, 10e-6, 10e-7]
hidden_layers = [2, 3, 4, 5, 6, 7]
score = 0

n = len(learning_rate)*len(momentum)*len(regularization)*len(hidden_layers)
i = 0
f_lr, f_m, f_r, f_hl = 0, 0, 0, 0

#Grid Search
for lr in learning_rate:
  for m in momentum:
    for r in regularization:
      for h in hidden_layers:
        brain = MLPClassifier(
            hidden_layer_sizes=(h),
            learning_rate_init=lr,
            momentum=m,
            alpha=r,
            max_iter=100000
        )

        i += 1
        print(i, '/', n)
        new_score = np.mean(cross_val_score(brain, X, Y, cv=3))
        if new_score > score:
          score = new_score
          f_lr = lr
          f_m = m
          f_r = r
          f_hl = h

print('Learning rate: ', lr)
print('Momentum: ', m)
print('Regularization: ', r)
print('Hidden layers: ', hl)
print('Score: ', score)
```
