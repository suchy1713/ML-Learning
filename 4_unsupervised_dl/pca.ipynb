```
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

#generate some data
X, y = make_classification(n_samples=1000, n_features=20, n_redundant=10, n_informative=5, n_classes=2)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```
import numpy as np
import matplotlib.pyplot as plt

#compute the X covariance and eigens
covX = np.cov(X_train.T)
lambdas, Q = np.linalg.eigh(covX)

#sort lambdas and eigenvectors
idx = np.argsort(-lambdas)
lambdas = lambdas[idx]
Q = Q[:, idx]

#transform initial X
Z = X_train.dot(Q)

#plot data in 2 dims
plt.scatter(Z[:, 0], Z[:, 1], c=y_train, alpha=0.3)
plt.show()
```

```
#plot variance of each component
plt.plot(lambdas)
plt.show()

#plot cumulative variance
plt.plot(np.cumsum(lambdas)/np.sum(lambdas))
plt.show()
```
